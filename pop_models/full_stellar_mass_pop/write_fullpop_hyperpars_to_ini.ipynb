```python
import popsummary
import numpy as np
```

```python
pdb_file = '../../../data/GWTC-4_pop/analyses_AllCBC/AllCBC_FullPop.h5'
res = popsummary.popresult.PopulationResult(fname=pdb_file)

logl = res.get_hyperparameter_samples(hyperparameters=["log_likelihood"])
maxl = np.argmax(logl)

def get(name):
    return res.get_hyperparameter_samples(
        hyperparameters=[name]
    )[maxl]

hyperparams = dict(
    alpha_1=get("alpha_1"), alpha_dip=get("alpha_dip"), alpha_2=get("alpha_2"),
    NSmin=get("NSmin"), NSmax=get("NSmax"),
    BHmin=get("BHmin"), BHmax=get("BHmax"),
    UPPERmin=get("UPPERmin"), UPPERmax=get("UPPERmax"),
    n0=get("n0"), n1=get("n1"),
    n2=get("n2"), n3=get("n3"),
    n4=get("n4"), n5=get("n5"),
    mix1=get("mix1"), mu1=get("mu1"), sig1=get("sig1"),
    mix2=get("mix2"), mu2=get("mu2"), sig2=get("sig2"),
    A=get("A"), A2=get("A2"),
    beta_pair_1=get("beta_pair_1"), beta_pair_2=get("beta_pair_2"),
    mbreak=get("mbreak")
)
```

```python
hyperparams
```

```python
with open('pdb_hyperpar_defaults_test.ini', 'w') as f:
    f.write("[hyperparameters]\n")
    for key, value in hyperparams.items():
        f.write(f"{key} = {value[0]}\n")
```
