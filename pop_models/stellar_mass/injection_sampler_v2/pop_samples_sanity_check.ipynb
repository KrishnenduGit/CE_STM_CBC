```python
import h5py
import pylab as plt
import os
```

```python
data_dir = './'
```

## BBH

```python
samples_file = os.path.join(data_dir, 'pop_m_FP4_spin_custom_fref_5.0_z_MD_lmr_130_BBH_34K_samples_v2.h5')
```

```python
with h5py.File(samples_file, 'r') as f:
    print(f.keys())
    samples = {key:val[()] for key, val in f.items() if key != 'config'}
```

```python
import popsummary
import numpy as np

pdb_file = '/home/divyajyoti/ACADEMIC/Projects/CE_STM/CBC/data/GWTC-4_pop/analyses_BBH/BBHSpin_MagTruncnormIidTiltIsotropicTruncnormNid.h5'
res = popsummary.popresult.PopulationResult(fname=pdb_file)

chi_grid_default, chi_rates_default   = res.get_rates_on_grids('a')
cost_grid_default, cost_rates_default = res.get_rates_on_grids('cos_tilt')

a_paper = chi_grid_default[0]
prob_a_paper = np.median(chi_rates_default, axis=0)
cost_paper = cost_grid_default[0]
prob_cost_paper = np.median(cost_rates_default, axis=0)
```

```python
plt.plot(a_paper, prob_a_paper, label='GWTC-4')
plt.hist(samples['a_1'], label='a_1', bins=80, histtype='step', density=True);
plt.hist(samples['a_2'], label='a_2', bins=80, histtype='step', density=True);
plt.legend()
```

```python
plt.plot(cost_paper, prob_cost_paper, label='GWTC-4')
plt.hist(samples['cos_tilt_1'], label='cos_tilt_1', bins=80, histtype='step', density=True);
plt.hist(samples['cos_tilt_2'], label='cos_tilt_2', bins=80, histtype='step', density=True);
plt.legend()
```
