# BBHs

```python
pop_dict = {}

pop_dict['m1_src'] = f['mass_1_source'][:]
pop_dict['m2_src'] = f['mass_2_source'][:]
pop_dict['z'] = f['redshift'][:]
pop_dict['dL'] = f['luminosity_distance'][:]/1000 # Convert to Gpc

pop_dict['chi1z'] = f['spin_1z'][:]
pop_dict['chi2z'] = f['spin_2z'][:]

pop_dict['psi'] = f['psi'][:]
pop_dict['thetaJN'] = f['theta_jn'][:]

pop_dict['theta'] = np.pi/2. - f['dec'][:]
pop_dict['phi'] = f['ra'][:]
pop_dict['ra'] = f['ra'][:]
pop_dict['dec'] = f['dec'][:]

pop_dict['Phicoal'] = f['phase'][:]
pop_dict['tcoal'] = GPSt_to_GMST_alt(f['geocent_time'][:])

# Computing Mc and eta in detector frame
m1_det = pop_dict['m1_src']*(1.+pop_dict['z'])
m2_det = pop_dict['m2_src']*(1.+pop_dict['z'])
pop_dict['Mc'] = ((m1_det*m2_det)**(3./5.))/((m1_det+m2_det)**(1./5.))
pop_dict['eta'] = (m1_det*m2_det)/((m1_det+m2_det)*(m1_det+m2_det))

pop_dict.keys()
```

```python
# Saving samples in a h5 file
import h5py
fname = 'BBH_cat_1yr.h5'
with h5py.File(fname, 'w') as out:
                
    def cd(n, d):
        d = np.array(d)
        out.create_dataset(n, data=d, compression='gzip', shuffle=True)
    
    for key in pop_dict.keys():
        cd(key, pop_dict[key])
```
